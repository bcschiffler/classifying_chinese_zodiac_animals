# Classifying Animals in the Chinese Zodiac

## Based on fastai [lesson 2](https://nbviewer.jupyter.org/github/fastai/course-v3/blob/master/nbs/dl1/lesson2-download.ipynb)

```python
from fastai.vision import *
```

### Downloading images for the 12 Zodiac classes from Google as specified in the [lesson 2](https://nbviewer.jupyter.org/github/fastai/course-v3/blob/master/nbs/dl1/lesson2-download.ipynb) notebook

### Create directories

```python
folders = ['rat', 'ox', 'tiger', 'rabbit', 'dragon', 'snake', 'horse', 'sheep', 'monkey', 'rooster', 'dog', 'pig']

for folder in folders:
  
  filename = 'urls_' + folder + '.txt'

  path = Path('data/chinese_zodiac_animals')
  dest = path/folder
  dest.mkdir(parents=True, exist_ok=True)
  
  download_images(filename, dest, max_pics=200)
  
```

```python
for c in folders:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

```python
path.ls()
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=3, figsize=(7,8))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

## Train model

```python
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

### Achieving roughly 96% accuracy in the first phase

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1_chinese_zodiac_animals')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

### The second phase does not further increase accuracy

```python
learn.fit_one_cycle(3, max_lr=slice(1e-5,1e-3))
```

```python
learn.save('stage-2_chinese_zodiac_animals')
```

## Interpretation

```python
learn.load('stage-2_chinese_zodiac_animals');
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

## Investigating wrong predictions

```python
interp.plot_top_losses(16)
```
